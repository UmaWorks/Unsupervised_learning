```python
#from config import gkey
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import gmaps
# Configure gmaps
gmaps.configure(api_key="AIzaSyBWakW8_j5eGSM6ySec9-1VoUj9x8qM4mc")
from us import states
```

```python
df=pd.read_excel("orders.xlsx")
df=df.set_index(['order_id'])
df.head()
```

```python
df["customer_billing_zip"]=df["customer_billing_zip"].apply(lambda x:x.split("-")[0] if "-" in str(x) else x)
df.head()
```

```python
pivot_order_zipcode = pd.pivot_table(df,values='order_amount', index=['customer_billing_zip'] ,aggfunc=np.sum).reset_index()

top_Sales_zipcode =pivot_order_zipcode.sort_values(by ='order_amount',ascending = False).head(1000)
top_Sales_zipcode
```

```python
#Getting the latitude and longitude from the top zipcodes
import pgeocode
lats=[]
lons =[]
nomi = pgeocode.Nominatim('us')
for zips in top_Sales_zipcode['customer_billing_zip']:
    lat = nomi.query_postal_code(zips).latitude
    lon = nomi.query_postal_code(zips).longitude
    lats.append(lat)
    lons.append(lon)
    
lats    
top_Sales_zipcode['lats']= lats
top_Sales_zipcode['lons']= lons
top_Sales_zipcode = top_Sales_zipcode[top_Sales_zipcode['lats'].notnull()]
top_Sales_zipcode
```

```python
# Storing 'Lat' and 'Lng' into  locations 
locations = top_Sales_zipcode [['lats', 'lons']].astype(float)
locations
```

```python
# Converting total sales to float and store
total_sales = top_Sales_zipcode ['order_amount'].astype(float)
total_sales
```

```python
# Customizing the size of the figure
figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Assigning the marker layer to a variable
markers = gmaps.marker_layer(locations)
# Adding the layer to the map
fig.add_layer(markers)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=total_sales,dissipating=True,
                                )
                                 

fig.add_layer(heat_layer)

fig
#plt.savefig(os.path.join("word_cloud_for_Samples.jpg"), dpi=1000, format='jpg', bbox_inches='tight')
```

```python
#configuring the google map with zipcode marker and their total sales details
fig = gmaps.figure()

sales_layer = gmaps.symbol_layer(locations,fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=[f"Total Sales amount: ${sale}" for sale in total_sales]
)


fig = gmaps.figure()
fig.add_layer(sales_layer)

fig
```
